```python
dir_ = '../extract_test/mosaics'

import os
import rasterio
import matplotlib.pyplot as plt

filename = os.path.join(dir_, os.listdir(dir_)[1])

raster = rasterio.open(filename).read([10,9,8])
def normalize(x):
    return (x - x.min()) / (x.max() - x.min())

plt.figure(figsize=(10,10))
plt.imshow(normalize(raster.transpose(1,2,0)))
plt.colorbar()
plt.axis('off')
```
